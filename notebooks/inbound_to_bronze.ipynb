##Leitura dos dados 
Lendo os dados da camada inbound

```scala
val path = "dbfs:/mnt/dados/inbound/dados_brutos_imoveis.json"
```

```scala
val dados = spark.read.json(path)
```

```scala
display(dados)
```

- Algumas colunas são dispensaveis na estrutura acima; necessita remoçao de colunas 

##Removendo Colunas

```scala
val dados_anuncio = dados.drop("imagens","usuario")
display(dados_anuncio)
```

##Criando coluna de identificação

```scala
import org.apache.spark.sql.functions.col
```

```scala
val df_bronze = dados_anuncio.withColumn("id",col("anuncio.id"))
display(df_bronze)
```

##Salvando na camada bronze
Salvando os dados obtidos na inbound e inicialmente tratados na camada bronze

```scala
val path_bronze = "dbfs:/mnt/dados/bronze/dataset_imoveis"
```

```scala
df_bronze.write.format("delta").mode(SaveMode.Overwrite).save(path_bronze)
```

```scala
display(dbutils.fs.ls("/mnt/dados/bronze/dataset_imoveis"))
```
